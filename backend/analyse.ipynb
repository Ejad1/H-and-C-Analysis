# Analyse et études des paroles du cantique *Hymnes & Cantiques*

## Extraction des paroles de chaque cantique

```python
from pdf2image import convert_from_path
import pytesseract
import os

# Chemin vers ton fichier PDF
pdf_path = "../data/HeC_A4.pdf"

# Convertir les pages du PDF en images
images = convert_from_path(pdf_path, dpi=300)

# Pour chaque page, faire l'OCR
for i, image in enumerate(images):
    text = pytesseract.image_to_string(image, lang='fra')
    print(f"--- Page {i + 1} ---")
    print(text)
    print("\n")
```

```python
from pdf2image import convert_from_path
import pytesseract
import os
import gc  # pour libérer la mémoire

# Chemin vers ton fichier PDF
pdf_path = "../data/HeC_A4.pdf"

# Créer un dossier pour stocker les images (facultatif)
os.makedirs("images_temp", exist_ok=True)

# Déterminer le nombre de pages (sans tout charger)
from PyPDF2 import PdfReader
num_pages = len(PdfReader(pdf_path).pages)

# OCR page par page
for i in range(1, num_pages + 1):
    try:
        print(f"Traitement de la page {i}...")
        
        # Convertir une seule page en image
        images = convert_from_path(pdf_path, dpi=200, first_page=i, last_page=i)
        image = images[0]

        # OCR
        text = pytesseract.image_to_string(image, lang='fra')

        # Affichage ou sauvegarde
        print(f"\n--- Page {i} ---\n{text}\n")

        # Facultatif : sauvegarder dans un fichier texte
        with open(f"ocr_page_{i}.txt", "w", encoding="utf-8") as f:
            f.write(text)

        # Nettoyage mémoire
        del image
        del images
        gc.collect()

    except Exception as e:
        print(f"❌ Erreur à la page {i} : {e}")
```
